Import pandas and matplot lib.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Load data set in as CSV.

```python
file_path = '../data/ADA_profile_simplified.csv'  
df = pd.read_csv(file_path)
df
```

Filter only for rows that are in the Greater Toronto Area.
Here we are using the Toronto CMA, which includes surrounding CMAs like Oshawa and Hamilton.

```python
gta_df = df[df["CMANAME"].str.contains("Toronto", case=False, na=False)]
gta_df
```

Create non-immigrant population data frame. Select relevant columns: area name (CSDNAME), total non-immigrant population (T1528), and ADA code (ADA_code).

```python
nonimmigrant_df = gta_df[["ADA_code", "CSDNAME", "T1528"]].copy()
nonimmigrant_df = gta_df.groupby("CSDNAME")["T1528"].sum().reset_index()
nonimmigrant_df = nonimmigrant_df.rename(columns={"T1528": "Non-Immigrant_Population"})
nonimmigrant_df = nonimmigrant_df.sort_values(by="Non-Immigrant_Population", ascending=False)
nonimmigrant_df
```

Create immigrant population data frame. Select relevant columns: area name (CSDNAME), total non-immigrant population (T1528), and ADA code (ADA_code).

```python
immigrant_df = gta_df[["ADA_code", "CSDNAME", "T1529"]].copy()
immigrant_df = gta_df.groupby("CSDNAME")["T1529"].sum().reset_index()
immigrant_df = immigrant_df.rename(columns={"T1529": "Immigrant_Population"})
immigrant_df = immigrant_df.sort_values(by="Immigrant_Population", ascending=False)
immigrant_df
```

Merge both data frames with CSDNAME.

```python
df = pd.merge(immigrant_df, nonimmigrant_df, on='CSDNAME')
df
```

Plot data.

```python
# Bar chart parameters
x = np.arange(len(df['CSDNAME']))  
width = 0.35  

# Create the plot
fig, ax = plt.subplots(figsize=(10, 10))
bars1 = ax.bar(x - width/2, df['Immigrant_Population'], width, label='Immigrant')
bars2 = ax.bar(x + width/2, df['Non-Immigrant_Population'], width, label='Non-Immigrant')

# Labels and titles
ax.set_xlabel('Area (CSDNAME) in GTA')
ax.set_ylabel('Population')
ax.set_title('Immigrant vs Non-Immigrant Population by Area')
ax.set_xticks(x)
ax.set_xticklabels(df['CSDNAME'], rotation=90)
ax.legend()
```
