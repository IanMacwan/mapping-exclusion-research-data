Import pandas and matplot lib.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load data set in as CSV.

```python
file_path = '../data/ADA_acs_file.csv'  
df = pd.read_csv(file_path)
df
```

Filter only for rows that are in the Greater Toronto Area.
Here we are using the Toronto CMA, which includes surrounding CMAs like Oshawa and Hamilton.

```python
gta_df = df[df["CMANAME"].str.contains("Toronto", case=False, na=False)]
gta_df
```

Group by Census Subdivision and sum walk/transit access metrics for primary and secondary educational facilities.

```python
gta_summary = gta_df.groupby('CSDNAME')[['walk_ef', 'public_ef']].sum()
gta_summary
```

Sort data to prepair for plotting.

```python
plot_data = gta_summary[['walk_ef', 'public_ef']].sort_values(by='walk_ef', ascending=False)
plot_data
```

Plot data.

```python
plt.figure(figsize=(16, 16))
plot_data.plot(kind='bar', width=0.8)
plt.title('Walkability and Transit Access to Primary and Secondary Educational Facilities by Census Subdivision in GTA', fontsize=15)
plt.xlabel('Census Subdivision', fontsize=12)
plt.ylabel('Accessibility Score', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Access Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
