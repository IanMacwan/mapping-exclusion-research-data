Import pandas and matplot lib.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load data set in as CSV.

```python
file_path = 'ADA_profile_simplified.csv'  
df = pd.read_csv(file_path)
df
```

Filter only for rows that are in the Greater Toronto Area.
Here we are using the Toronto CMA, which includes surrounding CMAs like Oshawa and Hamilton.

```python
gta_df = df[df["CMANAME"].str.contains("Toronto", case=False, na=False)]
gta_df
```

Select relevant columns: area name (CSDNAME), total non-immigrant population (T1528), and ADA code (ADA_code).

```python
nonimmigrant_df = gta_df[["ADA_code", "CSDNAME", "T1528"]].copy()
```

Add up all non-immigrant populations of duplicate area names. This is done because duplicate area names represent different sub populations of non-immigrants within the given area name. For example there are four rows that share the Orangeville area name, however, it can be noted that the 4 rows have distinct dissemination area codes (ADA_code), meaning they each represent a sub area of Orangeville. Thus to get the total non-immigrant population for Orangeville we would simply need to sum all sub non-immigrant populations i.e. all rows with area name Orangeville.

```python
nonimmigrant_df = gta_df.groupby("CSDNAME")["T1528"].sum().reset_index()
nonimmigrant_df = nonimmigrant_df.rename(columns={"T1528": "Non-Immigrant_Population"})
```

Sort the popoulation by descending order.

```python
nonimmigrant_df = nonimmigrant_df.sort_values(by="Non-Immigrant_Population", ascending=False)
```

Note the values below, showing the non-immigrant population of areas in the GTA.

```python
nonimmigrant_df
```

Plot values using matplotlib.

```python
plt.figure(figsize=(14, 8))
plt.bar(nonimmigrant_df["CSDNAME"], nonimmigrant_df["Non-Immigrant_Population"], color='red')
plt.xticks(rotation=45, ha='right')
plt.xlabel("Area (CSDNAME) in GTA")
plt.ylabel("Non-Immigrant Population")
plt.title("Non-Immigrant Population by Area in Greater Toronto Area (2021 Census)")
plt.tight_layout()
plt.show()
```
