Import pandas and matplot lib.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load data set in as CSV.

```python
file_path = 'ADA_acs_file.csv'  
df = pd.read_csv(file_path)
df
```

Filter only for rows that are in the Greater Toronto Area.
Here we are using the Toronto CMA, which includes surrounding CMAs like Oshawa and Hamilton.

```python
gta_df = df[df["CMANAME"].str.contains("Toronto", case=False, na=False)]
gta_df
```

Define the columns for overall walking and public transit access.

```python
walk_cols = ['walk_caf', 'walk_ef', 'walk_emp', 'walk_gs', 'walk_hf', 'walk_psef', 'walk_srf']
transit_cols = ['public_caf', 'public_ef', 'public_emp', 'public_gs', 'public_hf', 'public_psef', 'public_srf']
```

Group by Census Subdivision and sum all walk/transit access metrics.

```python
gta_summary = gta_df.groupby('CSDNAME')[[*walk_cols, *transit_cols]].sum()
gta_summary
```

Calculate total walkability and transit access per area.

```python
gta_summary['Walkability'] = gta_summary[walk_cols].sum(axis=1)
gta_summary['Transit Access'] = gta_summary[transit_cols].sum(axis=1)
```

Sort data to prepair for plotting.

```python
plot_data = gta_summary[['Walkability', 'Transit Access']].sort_values(by='Walkability', ascending=False)
plot_data
```

Plot data.

```python
plt.figure(figsize=(16, 16))
plot_data.plot(kind='bar', width=0.8)
plt.title('Overall Walkability and Transit Access by Census Subdivision in GTA', fontsize=16)
plt.xlabel('Census Subdivision', fontsize=12)
plt.ylabel('Accessibility Score', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Access Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
